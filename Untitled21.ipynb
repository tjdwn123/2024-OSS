# OSS(오픈 소스 소프트웨어) 보고서

## 1. 개요
오픈 소스 소프트웨어(Open Source Software, OSS)는 소스 코드가 공개되어 누구나 자유롭게 사용, 수정, 배포할 수 있는 소프트웨어를 의미합니다. OSS는 일반적으로 협업과 커뮤니티 기반으로 개발되며, 기술 발전과 창의성을 촉진하는 데 중요한 역할을 합니다.

### 주요 특징
- **자유로운 사용**: 사용에 제한이 없습니다.
- **수정 가능**: 소스 코드를 확인하고 필요에 따라 변경 가능.
- **배포 가능**: 수정된 버전을 배포할 수 있음.
- **공동 개발**: 커뮤니티 기반으로 개발되며, 기여자 간의 협력이 중요.

---

## 2. 역사
### 초기
- **1970년대**: 초기 소프트웨어는 하드웨어와 함께 무료로 배포되었습니다. 예: 유닉스의 초기 버전.
- **1980년대**: 소프트웨어가 상용화되면서 소스 코드 접근이 제한되기 시작.

### 전환점
- **1983년**: 리처드 스톨만(Richard Stallman)이 GNU 프로젝트를 시작하며 자유 소프트웨어 운동을 촉진.
- **1991년**: 리누스 토르발스(Linus Torvalds)가 리눅스 커널을 공개하며 오픈 소스의 상징적인 사례로 자리잡음.

### OSS 개념의 정립
- **1998년**: "오픈 소스"라는 용어가 등장. 오픈 소스 이니셔티브(Open Source Initiative, OSI)가 설립됨.

## 3. 대표적인 오픈 소스 소프트웨어
### 운영체제
- **Linux**: 리눅스 커널 기반의 운영체제.
- **FreeBSD**: 고성능 유닉스 계열 운영체제.

### 프로그래밍 언어
- **Python**: 폭넓게 사용되는 고급 프로그래밍 언어.
- **Ruby**: 심플함과 생산성을 강조하는 프로그래밍 언어.

### 데이터베이스
- **MySQL**: 가장 널리 사용되는 관계형 데이터베이스.
- **PostgreSQL**: 고급 기능을 제공하는 오픈 소스 데이터베이스.

### 웹 애플리케이션
- **WordPress**: 블로그 및 콘텐츠 관리 시스템(CMS).
- **Apache**: 가장 널리 사용되는 웹 서버 소프트웨어.

## 4. OSS 저작권
오픈 소스 소프트웨어의 저작권은 소프트웨어 라이선스에 의해 관리됩니다. 라이선스는 소프트웨어의 사용, 수정, 배포에 관한 권리를 정의합니다.

### 주요 OSS 라이선스
1. **GPL(General Public License)**:
   - 수정 및 배포는 가능하나, 수정한 소스 코드도 공개해야 함.
2. **MIT License**:
   - 사용, 수정, 배포가 자유로우며, 최소한의 제한만 존재.
3. **Apache License**:
   - 특허 사용에 대한 권리를 명시하며, 상용화도 가능.
4. **BSD License**:
   - 간단하고 유연하며, 상용 소프트웨어에도 사용 가능.

### 저작권 이슈
- 저작권 침해 문제: 라이선스 조건을 무시한 사용.
- 기여자의 권리: OSS에 기여한 개발자의 권리가 보호되어야 함.
- 상용화 문제: OSS를 기반으로 한 상용 소프트웨어의 개발과 관련된 논란.

## 5. 결론
OSS는 기술 발전과 소프트웨어 생태계의 다양성을 촉진하는 데 중요한 역할을 하고 있습니다. OSS를 제대로 활용하기 위해서는 각 라이선스의 조건을 이해하고 준수하는 것이 필수적입니다. 또한, OSS 커뮤니티에 기여함으로써 전 세계 개발자들과 협력할 수 있는 기회를 누릴 수 있습니다.
